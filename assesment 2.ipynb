```python
n= int(input("Enter a number:"))
def prime():
    flag=True
    if n<=1:
        flag=False
    else:
        for i in range(2,n):
            if n%i==0:
                flag=False
                break
    print("prime" if flag else "not prime")
prime()
```

```python
import re
pan=input("Enter pan:")
pattern=r'^[A-Z]{5}[0-9]{4}[A-Z]{1}'
if re.match(pattern,pan):
    print("valid pan")
else:
    print("not valid")
```

```python
37. Marks to Grade Conversion for N Students
python
CopyEdit

Task: Input multiple students and assign grades A-F.

Input: [85, 40, 77, 92, 60]  
Output: ['A', 'F', 'B', 'A', 'C']
```

```python
marks=[85, 40, 77, 92, 60]  
grades=[]
for m in marks:
    if m>=90:
        grades.append('A')
    elif m>=80:
        grades.append('B')
    elif m>=70:
        grades.append('C')
    elif m>=60:
        grades.append('D')
    else:
        grades.append('F')
print(grades)
```
